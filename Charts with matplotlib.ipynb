# barchart showing the blood sugar level of a population

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = [80, 100, 125, 200],  rwidth=0.95, color = "g")
```

# side by side histogram showing the blood sugar of men and women

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 88, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel("sugar range")
plt.ylabel('Total no of patients')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['green', 'orange'], 
         label=['men', 'women'], orientation="horizontal")
plt.legend()
```

# pie chart showing the expenses of a particular person

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, 
        autopct='%0.2f%%', shadow=True, explode=[0.2,0,0.5,0.8,0])
plt.show()
```

# How to save your work as PNG file

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, 
        autopct='%0.2f%%', shadow=True, explode=[0.2,0,0.5,0.8,0])
plt.savefig("piechart.png", bbox_inches ="tight", pad_inches =2)
```
