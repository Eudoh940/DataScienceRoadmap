# set day as index

```python
df = pd.read_csv("C:/Users/udoha/Desktop/pandasFolder/fillna2.csv",parse_dates=["day"])
df.set_index("day", inplace=True)
df
```

```python
new_df = df.fillna("new name")
new_df
```

# replace all NAN value with something meaningful

```python
new_df = df.fillna(0)
new_df
```

# fill each column with specific character or string using the fillna method

```python
new_df = df.fillna({
    'temperature': "0",
    "windspeed": 0,
    "event": "no event"
})
new_df
```

# fill up the missing spaces with ffill method rather than making the values drop so much

```python
new_df = df.fillna(method="ffill")
new_df
```

# fill up the missing spacess with bfill method. this one helps to fill up values from previous days

```python
new_df = df.fillna(method="bfill")
new_df
```

# using limit parameters

```python
new_df = df.fillna(method ="ffill", limit=1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

# drop any cell that has NA

```python
new_df = df.dropna()
new_df
```
