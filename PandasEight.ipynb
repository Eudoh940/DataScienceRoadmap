# Concatenation

```python
import pandas as pd

nigeria_weather = pd.DataFrame({
    "city": ["ikeja", "lekki", "mushin"],
    "temperature":[32, 45, 30],
    "humidity": [80, 60, 78]
})
nigeria_weather
```

```python
ghana_weather = pd.DataFrame({
    "city": ["accra", "kumasi", "obuasi"],
    "temperature":[37, 21, 34],
    "humidity": [70, 63, 67]
})
ghana_weather
```

```python
df = pd.concat([nigeria_weather, ghana_weather])
df
```

```python
df = pd.concat([nigeria_weather, ghana_weather], ignore_index=True)
df
```

```python
df = pd.concat([nigeria_weather, ghana_weather], keys=["nigeria", "ghana"])
df
```
